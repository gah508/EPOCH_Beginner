### (2) 2017년과 2022년 데이터를 비교하여, 서울특별시(시도명: "서울특별시") 내에서 위치 변경이 있었던 업소들을 찾으세요. 위치 변경 여부는 경도와 위도의 변화로 판단하고, 상호명과 변화된 위치(경도, 위도) 차이를 '경도차이', '위도차이' 열을 새로 생성하여 출력하세요. (경도차이, 위도차이는 절댓값으로 출력되지 않아도 됩니다.)
* Hint: 비교 연산자 사용 (|, =! 등)

```python
# 서울특별시 17년, 22년 필터링
ma17s = ma17[ma17['시도명'] == '서울특별시']
ma22s = ma22[ma22['시도명'] == '서울특별시']
```

```python
# 상가업소번호를 기준으로 병합
merg = pd.merge(ma17s, ma22s, on='상가업소번호', suffixes=('_17', '_22'))
```

```python
# 경도, 위도차이
merg['경도차이'] = merg['경도_22'] - merg['경도_17']
merg['위도차이'] = merg['위도_22'] - merg['위도_17']
loc = merg[(merg['경도차이'] != 0) | (merg['위도차이'] != 0)]
```

```python
result = loc[['상호명_17', '경도차이', '위도차이']]
result
```

## **2️⃣‘movie_data.csv’ 파일을 불러와 영화 데이터를 시각화하세요.**

### (0) 데이터를 불러오세요.

```python
file_path3 = '/content/drive/MyDrive/Colab Notebooks/EPOCH/Beginer/Reviewfile/movies.csv'
movi = pd.read_csv(file_path3)

movi
```

### (1) 아래와 같은 평점('score') 상위 10개 영화의 관객 수에 대한 수평 막대 그래프를 그리세요.
1-1번.png

```python
import matplotlib.pyplot as plt

# 탑텐
top10 = movi.sort_values(by='score', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top10['name'], top10['votes'], color='#0047AB')

plt.title('Top 10')
plt.xlabel('Audience Count')

plt.show()
```

### (2) 아래와 같이 연도('year')별 평균 예산('budget')과 수익('gross')을 선 그래프로 나타내세요.
(사용한 색상은 b, forestgreen, orange, skyblue 입니다.)

1-2번.png

```python
# 연도별 평균 예산과 수익 계산
year = movi.groupby('year')[['budget', 'gross']].mean()

plt.figure(figsize=(10, 6))

plt.plot(year.index, year['budget'], color='orange', label='Budget', linestyle='--', marker='d', linewidth=2)
plt.plot(year.index, year['gross'], color='forestgreen', label='Gross', linestyle='-', marker='x', linewidth=2)

plt.axhline(y=1e8, color='blue', linestyle=':', linewidth=2)

plt.title('Average Movie Budget & Gross by Year')
plt.xlabel('Year')
plt.ylabel('Average Budget & Gross')
plt.legend()

plt.show()
```
