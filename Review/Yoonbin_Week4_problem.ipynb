### (2) 2017ë…„ê³¼ 2022ë…„ ë°ì´í„°ë¥¼ ë¹„êµí•˜ì—¬, ì„œìš¸íŠ¹ë³„ì‹œ(ì‹œë„ëª…: "ì„œìš¸íŠ¹ë³„ì‹œ") ë‚´ì—ì„œ ìœ„ì¹˜ ë³€ê²½ì´ ìˆì—ˆë˜ ì—…ì†Œë“¤ì„ ì°¾ìœ¼ì„¸ìš”. ìœ„ì¹˜ ë³€ê²½ ì—¬ë¶€ëŠ” ê²½ë„ì™€ ìœ„ë„ì˜ ë³€í™”ë¡œ íŒë‹¨í•˜ê³ , ìƒí˜¸ëª…ê³¼ ë³€í™”ëœ ìœ„ì¹˜(ê²½ë„, ìœ„ë„) ì°¨ì´ë¥¼ 'ê²½ë„ì°¨ì´', 'ìœ„ë„ì°¨ì´' ì—´ì„ ìƒˆë¡œ ìƒì„±í•˜ì—¬ ì¶œë ¥í•˜ì„¸ìš”. (ê²½ë„ì°¨ì´, ìœ„ë„ì°¨ì´ëŠ” ì ˆëŒ“ê°’ìœ¼ë¡œ ì¶œë ¥ë˜ì§€ ì•Šì•„ë„ ë©ë‹ˆë‹¤.)
* Hint: ë¹„êµ ì—°ì‚°ì ì‚¬ìš© (|, =! ë“±)

```python
# ì„œìš¸íŠ¹ë³„ì‹œ 17ë…„, 22ë…„ í•„í„°ë§
ma17s = ma17[ma17['ì‹œë„ëª…'] == 'ì„œìš¸íŠ¹ë³„ì‹œ']
ma22s = ma22[ma22['ì‹œë„ëª…'] == 'ì„œìš¸íŠ¹ë³„ì‹œ']
```

```python
# ìƒê°€ì—…ì†Œë²ˆí˜¸ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë³‘í•©
merg = pd.merge(ma17s, ma22s, on='ìƒê°€ì—…ì†Œë²ˆí˜¸', suffixes=('_17', '_22'))
```

```python
# ê²½ë„, ìœ„ë„ì°¨ì´
merg['ê²½ë„ì°¨ì´'] = merg['ê²½ë„_22'] - merg['ê²½ë„_17']
merg['ìœ„ë„ì°¨ì´'] = merg['ìœ„ë„_22'] - merg['ìœ„ë„_17']
loc = merg[(merg['ê²½ë„ì°¨ì´'] != 0) | (merg['ìœ„ë„ì°¨ì´'] != 0)]
```

```python
result = loc[['ìƒí˜¸ëª…_17', 'ê²½ë„ì°¨ì´', 'ìœ„ë„ì°¨ì´']]
result
```

## **2ï¸âƒ£â€˜movie_data.csvâ€™ íŒŒì¼ì„ ë¶ˆëŸ¬ì™€ ì˜í™” ë°ì´í„°ë¥¼ ì‹œê°í™”í•˜ì„¸ìš”.**

### (0) ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¤ì„¸ìš”.

```python
file_path3 = '/content/drive/MyDrive/Colab Notebooks/EPOCH/Beginer/Reviewfile/movies.csv'
movi = pd.read_csv(file_path3)

movi
```

### (1) ì•„ë˜ì™€ ê°™ì€ í‰ì ('score') ìƒìœ„ 10ê°œ ì˜í™”ì˜ ê´€ê° ìˆ˜ì— ëŒ€í•œ ìˆ˜í‰ ë§‰ëŒ€ ê·¸ë˜í”„ë¥¼ ê·¸ë¦¬ì„¸ìš”.
1-1ë²ˆ.png

```python
import matplotlib.pyplot as plt

# íƒ‘í…
top10 = movi.sort_values(by='score', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top10['name'], top10['votes'], color='#0047AB')

plt.title('Top 10')
plt.xlabel('Audience Count')

plt.show()
```

### (2) ì•„ë˜ì™€ ê°™ì´ ì—°ë„('year')ë³„ í‰ê·  ì˜ˆì‚°('budget')ê³¼ ìˆ˜ìµ('gross')ì„ ì„  ê·¸ë˜í”„ë¡œ ë‚˜íƒ€ë‚´ì„¸ìš”.
(ì‚¬ìš©í•œ ìƒ‰ìƒì€ b, forestgreen, orange, skyblue ì…ë‹ˆë‹¤.)

1-2ë²ˆ.png

```python
# ì—°ë„ë³„ í‰ê·  ì˜ˆì‚°ê³¼ ìˆ˜ìµ ê³„ì‚°
year = movi.groupby('year')[['budget', 'gross']].mean()

plt.figure(figsize=(10, 6))

plt.plot(year.index, year['budget'], color='orange', label='Budget', linestyle='--', marker='d', linewidth=2)
plt.plot(year.index, year['gross'], color='forestgreen', label='Gross', linestyle='-', marker='x', linewidth=2)

plt.axhline(y=1e8, color='blue', linestyle=':', linewidth=2)

plt.title('Average Movie Budget & Gross by Year')
plt.xlabel('Year')
plt.ylabel('Average Budget & Gross')
plt.legend()

plt.show()
```
