```python

columns = ['Date', 'Price', 'Volume']
data = [
    ['2024-01-01', 100, 300],
    ['2024-01-02', 102, 280],
    ['2024-01-03', 98, 320],
    ['2024-01-04', 105, 400],
    ['2024-01-05', 107, 300]
]

df_stocks = pd.DataFrame(data=data, columns=columns)
df_stocks
```

```python
df_stocks.set_index('Date', inplace=True)
df_stocks
```

```python
df_stocks.loc['2024-01-02','Volume']
```

```python
cond = (df_stocks['Price'] >= 100) & (df_stocks['Price'] <= 105)

filtered_df = df_stocks[cond]
filtered_df
```

```python
cond2 = (df_stocks['Price'] >= 105) & (df_stocks['Volume'] < 300)
df_stocks.loc[cond2,'Volume'] = 350
filtered_df2 = df_stocks.loc[df_stocks['Volume']==350]
filtered_df2
```
