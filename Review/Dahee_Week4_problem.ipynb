* Hint: λΉ„κµ μ—°μ‚°μ μ‚¬μ© (|, =! λ“±)

```python
mkf_2017 = mk_2017[['μƒνΈλ…', 'μ‹λ„λ…','κ²½λ„', 'μ„λ„']]
mkf_2022 = mk_2022[['μƒνΈλ…', 'μ‹λ„λ…', 'κ²½λ„', 'μ„λ„']]

mks_2017 = mkf_2017[mkf_2017['μ‹λ„λ…'] == 'μ„μΈνΉλ³„μ‹']
mks_2022 = mkf_2022[mkf_2022['μ‹λ„λ…'] == 'μ„μΈνΉλ³„μ‹']

merged_data = pd.merge(left=mks_2017, right=mks_2022, on='μƒνΈλ…', suffixes=('_2017', '_2022'))

merged_data['κ²½λ„μ°¨μ΄'] = merged_data['κ²½λ„_2022'] - merged_data['κ²½λ„_2017']
merged_data['μ„λ„μ°¨μ΄'] = merged_data['μ„λ„_2022'] - merged_data['μ„λ„_2017']

location_changes = merged_data[(merged_data['κ²½λ„μ°¨μ΄'] != 0) | (merged_data['μ„λ„μ°¨μ΄'] != 0)]

location_changes[['μƒνΈλ…', 'κ²½λ„μ°¨μ΄', 'μ„λ„μ°¨μ΄']]
```
