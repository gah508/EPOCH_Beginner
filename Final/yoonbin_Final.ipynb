### (3) ÏïÑÎûòÏôÄ Í∞ôÏùÄ ÏàòÎ©¥ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÎÑ§ Í∞ÄÏßÄ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨ÏÑ∏Ïöî.
* colors = ['gold', 'gray','darkgray', 'silver', 'lightgray', 'whitesmoke'] & 'coral' & 'purple' & 'green'

image.png

```python
import matplotlib.pyplot as plt
import numpy as np

# ÏàòÎ©¥ Ïßà Î∂ÑÌè¨
lb = ['4', '5', '6', '7', '8', '9']
sz = [1, 3, 30, 20, 29, 19]
cl = ['lightgray', 'gray', 'darkgray', 'silver', 'yellow', 'lightgray']
ex = (0, 0, 0.1, 0, 0, 0)

# ÏàòÎ©¥ Ïßà ÌèâÍ∞Ä
slp = np.array([4, 5, 6, 7, 8, 9])
cnt = np.array([5, 10, 105, 80, 110, 70])

# ÎÇòÏù¥
age = np.arange(25, 61)
dur = np.random.uniform(5, 8, len(age))

# ÏàòÎ©¥ Ïßà Ï∂îÏÑ∏
q_by_age = np.random.uniform(4, 9, len(age))
```

```python
# Í∑∏Î¶º ÏÉùÏÑ±
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# ÌîåÎ°Ø 1
axs[0, 0].pie(sz, explode=ex, labels=lb, colors=cl, autopct='%1.1f%%', startangle=90)
axs[0, 0].axis('equal')
axs[0, 0].set_title('Distribution of Quality of Sleep')

# ÌîåÎ°Ø 2
axs[0, 1].bar(age, dur, color='orange')
axs[0, 1].set_xlabel('Age')
axs[0, 1].set_ylabel('Sleep Duration (Hours)')
axs[0, 1].set_title('Average Sleep Duration by Age')

# ÌîåÎ°Ø 3
axs[1, 0].bar(slp, cnt, color='purple')
axs[1, 0].set_xlabel('Quality of Sleep')
axs[1, 0].set_ylabel('Count')
axs[1, 0].set_title('Counts of Quality of Sleep Ratings')

# ÌîåÎ°Ø 4
axs[1, 1].plot(age, q_by_age, color='green', marker='o')
axs[1, 1].set_xlabel('Age')
axs[1, 1].set_ylabel('Quality of Sleep')
axs[1, 1].set_title('Quality of Sleep Trend by Age')

plt.tight_layout()

plt.show()
```
