```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
import seaborn as sns

# CSV ÌååÏùº Î∂àÎü¨Ïò§Í∏∞
sleep_data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
sleep_data.head()
```

### (1) ÏÑ±Î≥ÑÏóê Îî∞Î•∏ ÌèâÍ∑† ÏàòÎ©¥ ÏãúÍ∞Ñ('Sleep Duration')ÏùÑ Ï∂úÎ†•ÌïòÏÑ∏Ïöî.

```python
sleep_data.groupby('Gender')['Sleep Duration'].mean()
```

### (2) Ïä§Ìä∏Î†àÏä§ ÏàòÏ§Ä('Stress Level')Ïù¥ 3 Ïù¥ÏÉÅÏù∏ ÏÇ¨ÎûåÎì§Ïùò ÏàòÎ©¥ Ïßà('Quality of Sleep') ÌèâÍ∑†ÏùÑ Íµ¨ÌïòÏÑ∏Ïöî.

```python
sleep_data.query("`Stress Level` >= 3")['Quality of Sleep'].mean()
```

### (3) ÏïÑÎûòÏôÄ Í∞ôÏùÄ ÏàòÎ©¥ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÎÑ§ Í∞ÄÏßÄ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨ÏÑ∏Ïöî.
* colors = ['gold', 'gray','darkgray', 'silver', 'lightgray', 'whitesmoke'] & 'coral' & 'purple' & 'green'

image.png

```python
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ Î∞è ÏÉâÏÉÅ ÏÑ§Ï†ï
colors = ['gold', 'gray', 'darkgray', 'silver', 'lightgray', 'whitesmoke']
highlight_colors = ['coral', 'purple', 'green']
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 1. Distribution of Quality of Sleep (ÌååÏù¥ Ï∞®Ìä∏)
quality_counts = sleep_data['Quality of Sleep'].value_counts()
axs[0, 0].pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', colors=colors)
axs[0, 0].set_title('Distribution of Quality of Sleep')

# 2. Average Sleep Duration by Age (ÎßâÎåÄ Í∑∏ÎûòÌîÑ)
avg_sleep_by_age = sleep_data.groupby('Age')['Sleep Duration'].mean()
axs[0, 1].bar(avg_sleep_by_age.index, avg_sleep_by_age.values, color='coral')
axs[0, 1].set_title('Average Sleep Duration by Age')
axs[0, 1].set_xlabel('Age')
axs[0, 1].set_ylabel('Sleep Duration (Hours)')

# 3. Counts of Quality of Sleep Ratings (ÎßâÎåÄ Í∑∏ÎûòÌîÑ)
axs[1, 0].bar(, quality_counts, color='coral')
axs[1, 0].set_title('Counts of Quality of Sleep Ratings')
axs[1, 0].set_xlabel('Quality of Sleep')
axs[1, 0].set_ylabel('Count')

# 4. Quality of Sleep Trend by Age (ÏÑ† Í∑∏ÎûòÌîÑ)
axs[1, 1].plot(sleep_data['Age'], sleep_data['Quality of Sleep'], marker='o', linestyle='-', color='green')
axs[1, 1].set_title('Quality of Sleep Trend by Age')
axs[1, 1].set_xlabel('Age')
axs[1, 1].set_ylabel('Quality of Sleep')


plt.tight_layout()
plt.show()
```
