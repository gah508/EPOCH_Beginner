```python
# 예제
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('X-Label') # xlabel(), ylabel() 함수에 문자열 입력
plt.ylabel('Y-Label')
plt.show()
```

### 2. 여백 지정하기
>**`labelpad`** 파라미터: 축 레이블의 **여백(Padding)**을 지정

```python
# 예제
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('X-Label', labelpad=15)
plt.ylabel('Y-Label', labelpad=20)
plt.show()
```

### 3. 폰트 설정하기

```python
# 폰트 스타일 편리하게 재사용 가능
import matplotlib.pyplot as plt

font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size':14 }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'}

plt.plot([1,2,3,4], [2,3,5,10])
plt.xlabel('X-Axis', labelpad=15, fontdict=font1) # family: 글씨체,
plt.ylabel('Y-Label', labelpad=20, fontdict=font2)
plt.show()
```

## CH 05. Matplotlib 축 범위 지정하기

> **`xlim()`**: X축이 표시되는 범위를 지정하거나 반환

> **`ylim()`**: Y축이 표시되는 범위를 지정하거나 반환

> **`axis()`**: X,Y축이 표시되는 범위를 지정하거나 반환

### 1. 기본 사용 - **`xlim()`**, **`ylim()`**
>**`xlim(), ylim()`** 함수의 min, max값을 입력하거나 리스트, 튜플 형태로 입력

>*입력값이 없으면 데이터에 맞게 자동으로 범위를 지정

```python
# 예제
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [2,3,5,10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim([0, 5]) # X축의 범위: [xmin, xmax]
plt.ylim([0, 20]) # Y축의 범위: [ymin, ymax]

plt.show()
```

### 2. 기본사용 - **`axis()`**
>**`axis()`** 함수에 입력한 리스트(튜플)은 반드시 네 개의 값이 있어야 함
>
>(xmin, xmax, ymin, ymax)

>*입력값이 없으면 데이터에 맞게 자동으로 범위를 지정

```python
# 예제
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [2,3,5,10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0, 5, 0, 20]) # X,Y축의 범위: [xmin, xmax, ymin, ymax]

plt.show()
```

### 3. 옵션 지정하기

```python
# 예제
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [2,3,5,10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis('square') # X,Y축의 범위를 동일하게 지정

plt.show()
# aixs() 옵션: 'on' | 'off' | 'equal' | 'scaled' | 'tight' | 'auto' | 'normal' | 'image' | 'square'
```

### 4. 축 범위 얻기
>**`xlim(), ylim(), axis()`** 함수: 그래프 영역에 표시되는 X,Y축의 범위를 반환

```python
# 예제
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [2,3,5,10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

x_range, y_range = plt.xlim(), plt.ylim()
print(x_range, y_range)

axis_range = plt.axis('scaled')
print(axis_range)

plt.show()
# X축은 0.85에서 4.15, Y축은 1.6에서 10.4 범위로 표시됨
```

## CH 06. Matplotlib 선 종류 지정하기

### 1. 기본 사용
> 포맷 문자열 사용

```python
# 예제
import matplotlib.pyplot as plt

plt.plot([1,2,3], [4,4,4], '-', color='C0', label='Solid')
plt.plot([1,2,3], [3,3,3], '--', color='C0', label='Dashed')
plt.plot([1,2,3], [2,2,2], ':', color='C0', label='Dotted')
plt.plot([1,2,3], [1,1,1], '-.', color='C0', label='Dash-Dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

### 2. linestyle 지정하기
>**`plot()`** 함수에 **`linestyle`** 파라미터 값 지정

```python
# 예제
import matplotlib.pyplot as plt

plt.plot([1,2,3], [4,4,4], linestyle='solid', color='C0', label="'solid'")
plt.plot([1,2,3], [3,3,3], linestyle='dashed', color='C0', label="'dashed'")
plt.plot([1,2,3], [2,2,2], linestyle='dotted', color='C0', label="'dotted'")
plt.plot([1,2,3], [1,1,1], linestyle='dashdot', color='C0', label="'dashdot'")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.tight_layout()
plt.show()
```

### 3. 튜플 사용하기

>e.g. (0,(1,1)) = 'dotted'

>e.g. (0, (5,5)) = 'dashed'

>e.g. (0, (3,5,1,5)) = 'dashdotted'

```python
# 예제
import matplotlib.pyplot as plt

plt.plot([1,2,3], [4,4,4], linestyle=(0,(1,1)), color='C0', label='(0,(1,1))')
plt.plot([1,2,3], [3,3,3], linestyle=(0,(1,5)), color='C0', label='(0,(1,5))')
plt.plot([1,2,3], [2,2,2], linestyle=(0,(5,1)), color='C0', label='(0,(5,1))')
plt.plot([1,2,3], [1,1,1], linestyle=(0,(3,5,1,5)), color='C0', label='(0,(3,5,1,5))')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
```

### 4. 선 끝 모양 지정하기

>**`plot()`** 함수의 **`solid_capstyle, dahs_capstyle`**를 사용해 선의 끝 모양 지정

```python
# 예제
import matplotlib.pyplot as plt

plt.plot([1,2,3], [4,4,4], linestyle='solid', linewidth=10,
         solid_capstyle='butt', color='C0', label='solid+butt') # 'butt': 뭉뚝한 끝 모양
plt.plot([1,2,3], [3,3,3], linestyle='solid', linewidth=10,
         solid_capstyle='round', color='C0', label='solid+round') # 'round': 둥근 끝 모양

plt.plot([1,2,3], [2,2,2], linestyle='dashed', linewidth=10,
         dash_capstyle='butt', color='C1', label='dahsed+butt')
plt.plot([1,2,3], [1,1,1], linestyle='dashed', linewidth=10,
         dash_capstyle='round', color='C1', label='dashed+round')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
```

## CH 09. Matplotlib 그래프 영역 채우기
>**`fill_between()`**: 두 수평 방향의 곡선 사이를 채움

>**`fill_betweenx()`**: 두 수직 방향의 곡선 사이를 채움

>**`fill()`**: 다각형 영역을 채움


### 1. 기본 사용 - **`fill_between()`**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=0.5) #(x[1],y[1]), (x[2],y[2]), (x[1],0), (x[1],0)을 잇는 영역
plt.show()
```

### 2. 기본 사용 - **`fill_betweenx()`**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5) #(x[2],y[2]), (x[3],y[3]), (0, y[2]), (0, y[3])을 잇는 영역
plt.show()
```

### 3. 두 그래프 사이 영역 채우기

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,4,8]

plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y1[1:3], y2[1:3],color='lightgray', alpha=0.5) # (x[1], y[1]), (x[1], y[2]), (x[2], y[1]), (x[2], y[2])을 잇는 영역
plt.show()
```

### 4. 다각형 영역 채우기 - **`fill()`**

### 3. 눈금 스타일 지정하기
> **`tick_params()`** 함수: 눈금의 스타일을 다양하게 설정

**`axis`**: 설정이 적용될 축을 지정 {‘x’, ‘y’, ‘both’}

**`direction`**: ‘in’, ‘out’으로 설정하면 눈금이 안/밖으로 표시 {‘in’, ‘out’, ‘inout’}

**`length`**: 눈금의 길이를 지정

**`pad`**: 눈금과 레이블과의 거리를 지정

**`labelsize`**: 레이블의 크기를 지정

**`labelcolor`**: 레이블의 색상을 지정

**`top/bottom/left/right`**: True/False로 지정하면 눈금이 표시될 위치를 선택

**`width`**:눈금의 너비를 지정

**`color`**: 눈금의 색상을 지정

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', ''])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.tick_params(axis='x', direction='in', length=3, pad=6, labelsize=14, labelcolor='green', top=True)
plt.tick_params(axis='y', direction='inout', length=10, pad=15, labelsize=12, width=2, color='r')

plt.show()
```

### 3. 폰트 지정하기
>**`fontdict`** 파라미터에서 딕셔너리 형태로 폰트 스타일을 설정

>**`fontsize`**는 포인트 단위의 숫자를 입력하거나 ‘smaller’, ‘x-large’ 등의 상대적인 설정

>**`fontweight`**: {‘normal’, ‘bold’, ‘heavy’, ‘light’, ‘ultrabold’, ‘ultralight’}



```python
# 예제
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.title('Graph Title', loc='right', pad=20)

title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
plt.title('Graph Title', fontdict=title_font, loc='left', pad=20)

plt.show()
```

### 4. 타이틀 얻기
>**`plt.title()`**함수: 타이틀을 나타내는 Matplotlib text 객체를 반환

>**`get_position()`**과 **`get_text()`** 메서드를 사용해 텍스트 위치와 문자열 얻기

```python
# 예제
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
title_right = plt.title('Graph Title', loc='right', pad=20)

title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
title_left = plt.title('Graph Title', fontdict=title_font, loc='left', pad=20)

print(title_left.get_position())
print(title_left.get_text())

print(title_right.get_position())
print(title_right.get_text())

plt.show()
```

## CH 15. Matplotlib 수평선/수직선 표시하기

**`axhline()`**: 축을 따라 수평선을 표시

**`axvline()`**: 축을 따라 수직선을 표시

**`hlines()`**: 지정한 점을 따라 수평선을 표시

**`vlines()`**: 지정한 점을 따라 수직선을 표시

### 1. 수평선 그리기 - **`axhline(), hlines()`**
**`axhline()`**: 첫번째 인자(y값으로서 수평선의 위치)/ 두,세번재 인자(xmin, xmax) 0에서 1사이의 값 입력

**`hlines()`**: y, xmin,xmax 순서대로 입력 -> 점(xmin,y)에서 점(xmax,y)를 따라 수평선 표시

```python
import matplotlib.pyplot as plt

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)

plt.show()
```

### 2. 수평선 그리기 - **`axvline(), vlines()`**
**`axvline()`**: 첫번째 인자(x값으로서 수직선의 위치)/ 두,세번재 인자(ymin, ymax) 0에서 1사이의 값 입력

**`vlines()`**: x, ymin,ymax 순서대로 입력 -> 점(x,ymin)에서 점(x,ymax)를 따라 수직선 표시

```python
import matplotlib.pyplot as plt

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)

plt.show()
```
